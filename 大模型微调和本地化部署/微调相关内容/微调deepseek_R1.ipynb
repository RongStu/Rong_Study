## 微调与LoRa

### 1、什么是微调(Fine-tuning)

&#160;&#160;微调（Fine-tuning）是指在已经训练好（Pre-trained）的大模型基础上，使用特定的数据集进一步训练，使其适应特定任务或领域的全过程。


### 2、什么是Lora(Low rank adaptation)

&#160;&#160;lora是一种高效的模型微调技术，传统的全量微调需要更新模型的所有参数，需要巨大的显存和计算资源，而**Lora**采取了更加聪明的策略：
- 冻结原模型：不修改原来的模型参数，保证模型的底座能力不丢失
- 增加旁路：在模型中插入少量的‘适配器’参数进行训练，即 w = w0 + delta*w = w0+B*A
